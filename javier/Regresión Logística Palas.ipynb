```python
import pandas as pd
import numpy as np
df = pd.read_excel("C://Users//javie//Downloads//CMP//OperTot - PowerView_ARCH1.xlsx", ) 
#cambiar la dirección del excel para que ejecute correctamente
df.head()
```

```python
#Creación de variables de las fechas para que sean cuantitativas

import datetime
#df_palas["shiftdate"] = df_palas["shiftdate"].apply(lambda x: datetime.datetime.strptime((x), '%Y-%d-%m').date())
#df_palas['dia'] = df_palas["shiftdate"].apply(lambda x: datetime.datetime.strptime(x, '%d').date())
#df_palas['mes'] = df_palas["shiftdate"].apply(lambda x: datetime.datetime.strptime(x, '%m').date())
#df_palas['anio'] = df_palas["shiftdate"].apply(lambda x: datetime.datetime.strptime(x, '%Y').date())
def dia(fecha):
  aux = str(fecha.date()).split('-')
  dia = str(aux[2])
  return dia

def mes(fecha):
  aux = str(fecha.date()).split('-')
  mes = str(aux[1])
  return mes

def ano(fecha):
  aux = str(fecha.date()).split('-')
  ano = str(aux[0])
  return ano
```

```python
#Preparación del dataset para la creación del modelo de regresión logística con las cargas de palas
#Primero se debe crear el dataset correspondiente sacando las columnas de interés

#Primero agregamos al dataset la separación de las fechas
df['Dia'] = df.apply(lambda row : dia(row['shiftdate']),axis=1)
df['Mes'] = df.apply(lambda row : mes(row['shiftdate']),axis=1)
df['Año'] = df.apply(lambda row : ano(row['shiftdate']),axis=1)

#Se filtra para obtener solo las palas que tienen totaltons mayor que 0 y los tiempos efectivos (tmcat01) mayor que 0
df_palas = df[ ( df['unit'] == 'Pala' ) & ( df['totaltons'] !=0.0 ) & ( df['tmcat01']!=0.0 ) ]

#Se crean nuevas variables con los datos convertidos a kilotoneladas y horas de tiempo productivo
df_palas['new_totaltons'] = df['totaltons']/1000.0
df_palas['new_tmcat01'] = df['tmcat01']/3600.0

#creación de la variable de de productividad
df_palas['tons_per_time'] = df_palas['totaltons']/df['tmcat01']
df_palas['tons_per_time'] = df_palas['tons_per_time']*(3600/1000)
df_palas
```

```python
df_palas_new = df_palas.drop(columns = ['totaltons', 'tmcat01'] )
df_palas_new
```

```python
df_palas_new.columns
```

```python
#crear columna código en la que irán los unos y ceros según sea mayor o menor que 3.1 en la columna tons_per_time

def crearCodigos(dato):
    if dato>=3.1:
        return 1
    else:
        return 0

        
df_palas_new['codigo'] = df_palas_new['tons_per_time'].apply(crearCodigos)
df_palas_new
```

```python
#Aquí ya se tiene el dataset como se desea, con las columnas que se desean xD

df_palas_new = df_palas_new.drop(columns = ['ID', 'shiftindex', 'shift#', 'unit#','start', "loadcnt", "tmcat00", 'tmcat08', 'tmcat08', 'tmcat09','tmcat10','tmcat11','tmcat12','tmcat13','tmcat14', 'tmcat15','tmcat16', 'tmcat17','tmcat18','tmcat19', 'suma loadtime,idletime,spottime','resta tmcat01-suma', 'resta'])
df_palas_new
```

```python
#Preparación del X e y para la regresión logística

#Se deben eliminar las variables no cuantitativas para el buen uso del modelo
X = df_palas_new.drop(columns = ['eqmtid','shiftdate','tons_per_time', 'codigo', 'unit', 'locid', 'operid', 'pit', 'crew', 'new_totaltons', 'new_tmcat01'])
y = df_palas_new['codigo']
```

```python
X
```

```python
y
```

```python
#Se estandariza el X para un mejor resultado en el modelo
from sklearn.preprocessing import StandardScaler
X_std=StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

#Se segmenta el X e y en lo que se utilizará en train y test
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

#Se genera el modelo de Regresión Logística
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
#Coeficientes del modelo de Regresión Logística
logmodel.coef_
```

```python
#Deforma más visual, teniendo en cuenta las variables de X
coeff_df = pd.DataFrame(logmodel.coef_.reshape(16,1),X.columns,columns=['Coefficient'])
coeff_df
```

```python
#Intercepción con el eje de la curva generada
logmodel.intercept_
```

```python
#Predicción del modelo con el X_test
predictions = logmodel.predict(X_test)
```
